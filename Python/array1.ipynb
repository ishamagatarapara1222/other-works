## Array creation & Initial Setup


```python
import numpy as np

# Diya gaya array create karna
marks = np.array([45, 67, 89, 32, 90, 76, 55, 99, 41, 68])
print("Original Marks:", marks)
```

#### 1. Find total marks using sum()

```python
total_marks = marks.sum()  # ya np.sum(marks)
print("1. Total Marks:", total_marks)
```

### 2.Find average using mean()

```python
average_marks = marks.mean()
print("2. Average Marks:", average_marks)
```

### 3.Find highest and lowest marks

```python
highest= marks.max() 
lowest = marks.min()
print(f"3.Highest Marks: {highest}, Lowest Marks: {lowest}")
```

### 4.Sort Marks in ascending order

```python
sorted_marks = np.sort(marks)
print("4. Sorted Marks:", sorted_marks)
```

### 5. Display marks greater than 60

```python
marks_gt_60 = marks[marks > 60]
print("5. Marks greater than 60:", marks_gt_60)
```

### 6. Create mask for marks greater than 75

```python
mask_gt_75 = marks > 75
print("6. Mask for marks > 75:", mask_gt_75)
```

### 7. Reshape array into (2,5)

```python
reshaped_marks = marks.reshape(2, 5)
print("7. Reshaped Array (2x5):\n", reshaped_marks)
```

### 8. Add 5 grace marks using broadcasting

```python
grace_marks = marks + 5
print("8. Marks after adding 5 grace marks:", grace_marks)
```

### 9. Split array into 2 equal parts

```python
parts = np.split(marks, 2)
print("9. Split into 2 parts:")
print("   Part 1:", parts[0])
print("   Part 2:", parts[1])
```

### 10. Find index of mark 90

```python
index_of_90 = np.where(marks == 90)[0][0]
print("10. Index of mark 90:", index_of_90)
```
